```C++
#include "ConvexHull.h"

// 凸包检测
void ConvexHull::findConvexHull();
```
```mermaid
flowchart LR
    A[开始] --> B{点的数量 < 3}
    B -->|是| C[结束]
    B -->|否| D[寻找边界点]
    D --> E[初始化凸包结果]
    E --> F[找到最左边的点]
    F --> G[从最左边的点开始，逆时针遍历所有点]
    G --> H[将当前点添加到结果中]
    H --> I[遍历所有点，搜索下一个点 q]
    I --> J{点 i 比 q 更接近 p}
    J -->|是| K[更新 q = i]
    K --> L[将 q 设置为下一个 p]
    L --> M{p != 最左边的点}
    M -->|是| G
    M -->|否| N[结束]
```

```C++
#include "DecisionTree.h"

// 使用贪婪的自上而下的方法从给定数据集构建决策树
shared_ptr<Node> DecisionTree::buildTree(DecisionTree::CRITERION criterion, Table &table, const string &targetFields, const vector<string> &featureFields, int maxDepth);
```

```C++
#include "ShortestPath.h"

// 最短路径算法（Dijkstra）
vector<Place> ShortestPathPointToPoint::shortestPath(Place startPlace, Place endPlace);
```
```mermaid
flowchart LR
    A[开始] --> B[获取场景的数量]
    B --> C[初始化所有场景的最短路径和访问状态]
    C --> D[定义并初始化优先队列]
    D --> E{优先队列为空?}
    E -->|否| F[从优先队列中取出最短路径最小的场景]
    F --> G{已经找到终点?}
    G -->|是| H[结束循环]
    G -->|否| I[遍历当前场景的所有路径]
    I --> J{路径的起点或终点是当前场景，并且另一端还没有被访问过?}
    J -->|是| K[计算并更新最短路径]
    K --> L[将新的场景和最短路径添加到优先队列中]
    L --> E
    E -->|是| M[通过 previousPlace 数组回溯最短路径]
    M --> N[返回最短路径]
    N --> O[结束]
```

```C++
#include "ShortestPath.h"

// 计算最短路径
double ShortestPathPointToPoint::calcMinDistance(const vector<Place> &places);
```
```mermaid
flowchart LR
    A[开始] --> B{地点的数量少于2?}
    B -->|是| C[返回0.0]
    B -->|否| D[初始化最短距离为0]
    D --> E[遍历所有的地点，除了最后一个]
    E --> F[遍历所有的路径]
    F --> G{当前地点是路径的起点，且下一个地点是路径的终点?}
    G -->|是| H[将路径的长度添加到最短距离]
    H --> I[跳出内部循环，继续下一个地点]
    G -->|否| J{当前地点是路径的终点，且下一个地点是路径的起点?}
    J -->|是| K[将路径的长度添加到最短距离]
    K --> L[跳出内部循环，继续下一个地点]
    J -->|否| F
    I --> E
    L --> E
    E --> M[返回最短距离]
    M --> N[结束]
```
